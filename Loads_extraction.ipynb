```python
# -*- coding: utf-8 -*-
import pandas as pd
import re
import matplotlib.pyplot as pp
import numpy as np
import os
import glob
```

```python
#loadpattern=re.compile('\d*55\d*(018|019|020)')

load_files=glob.glob('*.pnt29b')

loadpattern=re.compile('\d{8}')
```

```python
# the function is only for one folder. To look thru the many folders one more change is required 
def loads(file):
    d=[]
    with open(file,'r') as l:
        for line in l:
            if loadpattern.search(line):
                s=line.split()
                pattern=re.compile('\d{9,30}')
                n=s.index(pattern.search(line).group())
                if 'fs' in s or 'fy' in s:
                    s[n+5:n+7]=[' '.join(s[n+5:n+7])]
                s[1:n]=[' '.join(s[1:n])]
                if len(s)!=17:
                    k=len(s)
                    s.extend(['non']*(17-k))
                    s[k-1],s[-1]=s[-1],s[k-1]
                for i,x in enumerate(s):
                    try:
                        s[i]=int(x)
                    except ValueError:
                        try:
                            s[i]=float(x)
                        except ValueError:  
                            pass
                d.append(s)
    return d

```

```python
data=[] 
  
for file in load_files:
    data.extend(loads(file))
```

```python
num=np.array(data, dtype='object')
```

```python
num=np.array(data, dtype='object')

inf=['ID', 'Location', 'Element','Fmax','Cond max','Fmin','Cond min', 'Mode', 'FRF',
     'GAG', 'DSO', 'DFRreq', 'MS', 'SEF', 'Stress Rating', 'Sc', 'DMF Set']

df=pd.DataFrame(num, columns=inf)

df['Fmax'] = df['Fmax'].astype(float)
df['Fmin'] = df['Fmin'].astype(float)
df['GAG'] = df['GAG'].astype(float)
df['DFRreq'] = df['DFRreq'].astype(float)
df['Element'] = df['Element'].astype(str)
df['Stress Rating'] = df['Stress Rating'].astype(float)
df['FRF'] = df['FRF'].astype(float)
#df['Sc'] = df['Sc'].astype(float)
df['DSO'] = df['DSO'].astype(int)
df['ID'] = df['ID'].astype(int)

df['R'] = df['Fmin']/df['Fmax']



```

```python
#Display Float with 3 digits after point
pd.options.display.float_format = '{:,.3f}'.format

#Create the Data set with cpecified Columns
ex=df[['Element','Location','DFRreq','Fmax','Fmin','R','Mode','GAG','DSO','SEF','Stress Rating','Sc','DMF Set']]

#Sort the Rows by ascending/descending
ex.sort_values(['Element','DSO'], axis=0, ascending=[True,False], inplace=False, kind='quicksort')
```

```python
#Create Excel file with values
ex.to_excel("Loads_Summary.xlsx",sheet_name='Sheet_name_1')

#Create html if needed

#ex.to_html('Loads.html')
```
